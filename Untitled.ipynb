## Column Meanings

### Column
Popularity: <br>
Danceability: a rating from 0 to 1 on how danceable a song is based on tempo, rhythm stability, beat strength <br>
              and overall regularity
Energy: a rating from 0 to 1 on energy level based on intensity and activity <br>
Key: the musical key the track is in. Example: 0 is C, 1 is C#, and so on <br>
Loudness: the average loudness (db) of a track from -60 to 0 with 0 as the loudest  <br>
Mode: modality of song - major is 1, minor is 0 <br>
Speechiness: describes prescence of spoken words in a track. A value close to 1 represents more speech in the track
(podcast, talk show) while a value closer to 0 represents music<br>
Acousticness: value indicates confidence if track is acoustic with 1 indicating the track is acoustic <br>
Instrumentalness: a rating from 0 to 1 on the amount of vocals in a song. A value closer to 1 indicates the song is more instrumental. <br>
Liveness: a rating from 0 to 1 predicting the probability the song was recorded with a live audience. A value > 0.8 indicates high likelihood <br>
Valence: a rating from 0 to 1 indicating the 'positivity' of a song. A value closer to 1 indicates higher positivity while a value closer to 0 indicates negativity (sad, depressed, angry) <br>
Tempo: measured in BPM (beats per minute) <br>
Duration_ms: length of the song <br>

More info can be found here https://www.kaggle.com/nadintamer/top-tracks-of-2017
